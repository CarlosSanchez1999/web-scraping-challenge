```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Scrape the website
html = browser.html

# Create a BeautifulSoup object from the scraped HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
table = soup.find_all('table', class_='table')
print (table)
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Heres an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
# Read in HTML tables into a DataFrame
df = pd.read_html('https://static.bc-edx.com/data/web/mars_facts/temperature.html')
mars_temperature_data = df[0]
```

```python
# Confirm DataFrame was created successfully
mars_temperature_data
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_temperature_data.dtypes
```

```python
# Change data types for data analysis
# Convert 'terrestrial_date' to datetime64[ns]
mars_temperature_data['terrestrial_date'] = pd.to_datetime(mars_temperature_data['terrestrial_date'])
```

```python
# Confirm type changes were successful by examining data types again
mars_temperature_data.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
unique_months = mars_temperature_data['month'].nunique()
unique_months
```

```python
# 2. How many sols (Martian days) worth of data are there?
sols_count = mars_temperature_data['sol'].count()
sols_count
```

```python
# 3. What is the average minimum temperature by month?
avg_min_temp = mars_temperature_data.groupby('month')['min_temp'].mean()
avg_min_temp
```

```python
# Plot the average minimum temperature by month
plt.bar(avg_min_temp.index, avg_min_temp.values, color='skyblue', edgecolor='black')
plt.title('Average Minimum Temperature by Month', fontsize=16) 
plt.xlabel('Month')  
plt.ylabel('Temperature in Celsius') 
plt.xticks(range(1, 13))  
plt.tight_layout()  
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
# Sort the average minimum temperature
sorted_avg_min_temp = avg_min_temp.sort_values()
# Plot the sorted average minimum temperature
plt.bar(sorted_avg_min_temp.index.astype(str), sorted_avg_min_temp.values, color='skyblue', edgecolor='black')
plt.title('Average Minimum Temperature by Month')
plt.xlabel('Month (Sorted by Temperature)')
plt.ylabel('Temperature in Celsius')
plt.tight_layout()
plt.show()
```

```python
# 4. What is the average pressure by month?
avg_min_pressure = mars_temperature_data.groupby('month')['pressure'].mean()
avg_min_pressure
```

```python
# Plot the average pressure by month
plt.bar(avg_min_pressure.index, avg_min_pressure.values, color='skyblue', edgecolor='black')
plt.title('Average Pressure by Month') 
plt.xlabel('Month', fontsize=14)  
plt.ylabel('Atmospheric Pressure') 
plt.xticks(range(1, 13))  
plt.tight_layout()  
plt.show()
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
# Sort the average pressure
sorted_avg_pressure = avg_min_pressure.sort_values()

# Plot the sorted average pressure
plt.bar(sorted_avg_pressure.index.astype(str), sorted_avg_pressure.values, color='skyblue', edgecolor='black')
plt.title('Average Pressure by Month') 
plt.xlabel('Month', fontsize=14)  
plt.ylabel('Atmospheric Pressure') 
plt.xticks(range(0, 13))  
plt.tight_layout()  
plt.show()
```

```python
# 5. How many terrestrial (Earth) days are there in a Martian year?
# Visually estimate the result by plotting the daily minimum temperature of each observation in the data set.
plt.plot(mars_temperature_data.index, mars_temperature_data['min_temp'])
plt.title('Minimum Temperature over Time', fontsize=16)
plt.xlabel('Number of Terrestrial Days')  
plt.ylabel('Minimum Temperature (C)')  

# Display the plot
plt.tight_layout()
plt.show()
```

#### Minimum Temperature

The graph shows the average minimum temperature on Mars by month, revealing seasonal changes. The coldest months are March (Month 3) and April (Month 4), with temperatures around -83C and -82C, while the warmest months are August (Month 8) and September (Month 9), with temperatures near -68C. The difference between the coldest and warmest months is about 15C, showing how seasons affect Mars' climate. The months between, like February (Month 2) and May (Month 5), show a transition from cold to warm. This pattern is caused by Mars' tilt and orbit, which create noticeable temperature shifts. Understanding these changes is important for missions on Mars, especially during colder months, and helps scientists learn more about the planet's climate.

#### Atmospheric Pressure

This graph shows the average atmospheric pressure on Mars by month. The highest pressure is observed in September (Month 9) and October (Month 10), reaching values above 900 Pa, while the lowest pressure occurs in May (Month 5) and June (Month 6), dropping to around 700 Pa. The difference between the highest and lowest pressures reflects seasonal atmospheric changes, likely driven by Mars axial tilt and solar heating, which cause gases in the atmosphere to condense or expand. These variations are critical for understanding Martian weather patterns and planning safe operations for missions. Atmospheric pressure affects activities like rover mobility and instrument calibration, making this data valuable for mission design.

#### Year Length

This graph shows the variation of minimum temperature on Mars over a span of terrestrial days, as recorded by Curiosity. The data reveals a clear cyclical pattern, indicating the influence of Martian seasons on temperature changes. The lowest temperatures occur around days 500 and 1500, corresponding to the coldest parts of the Martian winter, while the highest minimum temperatures are observed around days 250, 1000, and 1750, aligning with warmer seasons. The repeating cycle suggests that a Martian year spans approximately 687 Earth days, consistent with prior scientific estimates. This visualization highlights the dramatic temperature shifts caused by Mars thin atmosphere and its axial tilt, which drives significant seasonal variations. Understanding these patterns is essential for planning rover operations and studying the planet's climate system.
